```python
from sklearn.datasets import load_iris
```

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
iris = load_iris()
```

```python
X = iris.data
```

```python
X
```

```python
y = iris.target
```

```python
%matplotlib inline
plt.scatter(X[y==0, 0], X[y==0,1], color="green", label="Setosa")
plt.scatter(X[y==1, 0], X[y==1,1], color="red", label="Versicolor")
plt.scatter(X[y==2, 0], X[y==2,1], color="blue", label="Verginica")
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()
plt.show()
```

```python
%matplotlib inline
plt.scatter(X[y==0, 2], X[y==0,3], color="green", label="Setosa")
plt.scatter(X[y==1, 2], X[y==1,3], color="red", label="Versicolor")
plt.scatter(X[y==2, 2], X[y==2,3], color="blue", label="Verginica")
plt.title("Petal Length vs Petal Width")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3)
```

```python
km.fit_predict(X)
```

```python
km.inertia_
```

```python
error = []

for i in range(1, 11):
    k = KMeans(n_clusters=i)
    k.fit(X)
    error.append(k.inertia_)
```

```python
error
```

```python
plt.plot(range(1, 11), error, marker="*", markeredgecolor="red")
plt.show()
```
